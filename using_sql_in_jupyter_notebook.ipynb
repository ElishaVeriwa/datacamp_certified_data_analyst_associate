```python
%%sql
DROP TABLE IF EXISTS clerk_tasks;
```

```python
%%sql
-- Table for Outstanding Invoices
CREATE TABLE IF NOT EXISTS invoices (
    invoice_id VARCHAR(10) PRIMARY KEY, -- A unique key which identifies an invoice
    customer_name TEXT NOT NULL, -- Full name of the customer
    amount_due DECIMAL(10, 2), -- Total amount due from the customer
    due_date DATE, -- The due date of the total invoice amount
    status TEXT CHECK(status IN ('Current', 'Overdue', 'Paid')) 
);

-- Table for the Clerk's Task List
CREATE TABLE IF NOT EXISTS clerk_tasks (
    task_id INTEGER PRIMARY KEY AUTOINCREMENT,
    invoice_id VARCHAR(10), -- Invoice unique reference
    task_description TEXT,
    priority TEXT DEFAULT 'Medium',
    due_date DATE, -- The due date of the invoice
    is_completed BOOLEAN DEFAULT 0,
    created_at DATETIME DEFAULT CURRENT_TIMESTAMP,
    FOREIGN KEY (invoice_id) REFERENCES invoices(invoice_id),
    FOREIGN KEY(due_date) REFERENCES invoives(due_date)
);
```

The table containing all company invoices is now ready for use. As well all tasks relating to the Debtors Clerk have been facilitated through a table which are to be filter parameters of what the persona will have access to.

Of course it is paramount to verify whether the database structure is kept well aligned to the DBMS ISO standards. So to cut the story short, let's check if the how the data looks like.

```python
%%sql
SELECT 
    t.due_date AS task_deadline, -- Due date of amounts owed by customer
    i.customer_name, -- Customer name
    i.amount_due, -- The value due from the client
    t.task_description, -- As may be assigned/given as an instructive directive by their superior
    T.PRIORITY -- The level of urgency depending on the company cashflow
FROM clerk_tasks AS t
JOIN invoices AS i
    ON t.invoice_id = i.invoice_id
--WHERE t.is_completed = 0 -- Incase we would want to filter the data to be returned from the table
ORDER BY 
    CASE priority 
        WHEN 'Urgent' THEN 1 
        WHEN 'High' THEN 2 
        WHEN 'Medium' THEN 3 
        ELSE 4 
    END;
```

Here the priority of each task has been logically implemented within the database.

```python
%%sql
CREATE TABLE IF NOT EXISTS payments (
    payment_id INTEGER PRIMARY KEY AUTOINCREMENT,
    invoice_id VARCHAR(10),
    amount_paid DECIMAL(10, 2),
    payment_date DATE,
    payment_method TEXT,
    FOREIGN KEY (invoice_id) REFERENCES invoices(invoice_id)
);

-- Let's record a partial payment for INV-003
INSERT INTO payments (invoice_id, amount_paid, payment_date, payment_method)
VALUES ('INV-003', 1000.00, '2024-04-05', 'EFT');
```

```python
%%sql
SELECT 
    customer_name,
    amount_due AS original_amount,
    -- Calculate remaining balance after payments
    (amount_due - COALESCE((SELECT SUM(amount_paid) FROM payments WHERE payments.invoice_id = invoices.invoice_id), 0)) AS balance_remaining,
    CASE 
        WHEN julianday('now') - julianday(due_date) <= 0 THEN 'Current'
     WHEN julianday('now') - julianday(due_date) BETWEEN 1 AND 7 THEN '1-7 Days Overdue'
        WHEN julianday('now') - julianday(due_date) BETWEEN 7 AND 30 THEN '7-30 Days Overdue'
        WHEN julianday('now') - julianday(due_date) BETWEEN 31 AND 60 THEN '31-60 Days Overdue'
        ELSE '61+ Days Overdue (Critical)'
    END AS aging_bucket
FROM invoices
--WHERE status != 'Paid';
```

Instead, let's start off this as a VIEW table.

```python
%%sql
CREATE VIEW customer_ageing AS
SELECT 
    customer_name,
    amount_due AS original_amount,
    -- Calculate remaining balance after payments
    (amount_due - COALESCE((SELECT SUM(amount_paid) FROM payments WHERE payments.invoice_id = invoices.invoice_id), 0)) AS balance_remaining,
    CASE 
        WHEN julianday('now') - julianday(due_date) <= 0 THEN 'Current'
     WHEN julianday('now') - julianday(due_date) BETWEEN 1 AND 7 THEN '1-7 Days Overdue'
        WHEN julianday('now') - julianday(due_date) BETWEEN 7 AND 30 THEN '7-30 Days Overdue'
        WHEN julianday('now') - julianday(due_date) BETWEEN 31 AND 60 THEN '31-60 Days Overdue'
        ELSE '61+ Days Overdue (Critical)'
    END AS aging_bucket
FROM invoices;
--WHERE status != 'Paid';

SELECT *
FROM customer_ageing;
```
